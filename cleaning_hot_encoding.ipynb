Qualitative variables that keep repeating : type,subtype,cuisine_type,condition,heatingType

```python
params =["type","subtype","cuisine_type","condition","heatingType"]
```

```python
immozip_oneHot = immozip.copy()
```

```python
for param in params:
    test = param+'_test'
    immozip_oneHot[test] = pd.Categorical(immozip_oneHot[param])
    data_dummies = pd.get_dummies(immozip_oneHot[test],prefix=param+"_is")
    del immozip_oneHot[param]
    del immozip_oneHot[test]
    immozip_oneHot = pd.concat([immozip_oneHot,data_dummies],axis=1)
```
